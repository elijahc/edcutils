```python
from src.datasets import nhamcs, namcs
from src.datasets.nchs import topn
import seaborn as sns
```

```python
priv_op = namcs.load_data(year=[2010,2011,2012])
hosp_op = nhamcs.load_opd(year=[2010,2011,2012])
priv_peds = priv_op[priv_op.AGER.isin(['Under 15 years'])]
hosp_peds = hosp_op[hosp_op.AGER.isin(['Under 15 years'])]

priv_acute = priv_peds[priv_peds.MAJOR.isin(['New problem (less than 3 mos. onset)'])]
hosp_acute = hosp_peds[hosp_peds.MAJOR.isin(['New problem (less than 3 mos. onset)'])]
```

```python
def plot_topn(df,col,n=5):
    sns.set_context('talk')
    tn = topn(df,col,n=5)
    g = sns.countplot(y=col,order=tn.index.values,data=df)
```

```python
plot_topn(priv_acute,'RFV1')
```

```python
plot_topn(hosp_acute,'DIAG1')
```

```python
plot_topn(hosp_acute,'RFV1')
```
